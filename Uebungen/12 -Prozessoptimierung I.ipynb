# Aufgabe 12.3

Es wird angenommen, dass vier Faktoren den Geschmack eines Erfrischungsgetränks beeinﬂussen:  
A) die Art des Süssungsmittels  
B) das Verhältnis von Sirup zu Wasser   
C) Kohlensäuregehalt  
D) und Temperatur  
Jeder Faktor kann auf zwei Ebenen ausgeführt werden, was ein 24 Design ergibt. Bei jedem Durchlauf des Designs werden Proben des Getränks an eine Testgruppe Testpanel, bestehend aus 20 Personen, gegeben. Jeder Tester ordnet dem Getränk eine Punktzahl von 1 bis 10. Die Gesamtpunktzahl ist die Reaktionsvariable, und das Ziel ist es, eine Rezeptur zu ﬁnden, die die Gesamtpunktzahl maximiert. Es werden zwei Wiederholungen dieses Designs durchgeführt, und die Ergebnisse sind in der Tabelle dargestellt.

```python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
from patsy.contrasts import Treatment, Sum 
from statsmodels.stats.anova import anova_lm

df = pd.DataFrame({'Aa': np.tile(np.tile([-1, 1], 8),2), 
'Bb': np.tile(np.repeat([-1,1], [2,2]),8),
'Cc': np.tile(np.repeat([-1,1], [4,4]),4), 
'Dd': np.tile(np.repeat([-1,1], [8,8]),2)},
dtype='category')

df['Yy']=np.array([159,168,158,166,175,179,173,179,164,187,163,185,168,197,170,194,163,
175,163,168,178,183,168,182,159,189,159,191, 174, 199,174,198])

print(df)
```

a) Gibt es Variablen, die nicht relevant für die Gesamtpunktezahl ist (ohne Interaktion)?

```python
fit = ols("Yy ~ C(Aa, Sum)+C(Bb, Sum)+C(Cc, Sum)+C(Dd, Sum)", data=df).fit()
anova_lm(fit)
```

Die Variable B (Verhältnis Sirup Wasser) ist die einzige Variable, die nicht hochsigniﬁkant ist.

b) Gibt es statistisch signiﬁkante Interaktion? Können Sie diese begründen?  
Verwenden Sie α = 0.05 in den statistischen Tests.

```python
fit = ols("Yy ~ C(Aa, Sum)*C(Bb, Sum)*C(Cc, Sum)*C(Dd, Sum)", data=df).fit()
anova_lm(fit)
```

Die Variabeln A, C (und D?! --> gem. Lösung nicht!) weisen eine Signifikante Interaktion auf.
