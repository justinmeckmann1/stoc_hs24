```python
from scipy.stats import binom
import numpy as np
```

```python
# a) 
n = 30 # number of samples
k = 25 # number of correct samples
p = 0.95 # probability of success

W = binom.pmf(k=k, n=n, p=p)
print(f"W = {W}")
```

```python
# b) 
n = 30
k = 26 # number of correct samples
p = 0.95 # probability of success

W = 1-binom.cdf(k=k, n=n, p=p) # probability, that more then 26 samples are correct
print(f"W = {W}")
```

```python
# c) 
n = 30
k = 28 # max number of defective samples
p = 0.05 # probability of failure

Q = binom.cdf(k=k, n=n, p=p) # probability, that less then 28 samples are defective
print(f"Q = {Q}")
```

```python
# d --> min 29 correct samples 
n = 30
k = 29 # min number of correct samples
p = 0.95 # probability of success

W = 1-binom.cdf(k=k-1, n=n, p=p) # probability, that at least 29 samples are correct
print(f"W = {(W*100):.2f}%")
```

```python
# e) 
q = 0.05 # probability of failure
n = 30 # number of samples
p = 0.95 # probability of success

k = binom.ppf(q=q,n=n, p=p) # number of defective samples
print(f"k = {k}")
```
