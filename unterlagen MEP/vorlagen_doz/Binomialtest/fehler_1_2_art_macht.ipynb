## Einseitig

Wir beginnen mit einer Binomialverteilung mit $n=50$ und $\pi=0.6$

```python
%matplotlib inline

from scipy.stats import binom 
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)
import numpy as np
n = 50
p_0 = 0.6
```

## Fehler 1. Art

Signifikanzniveau $\alpha$

```python
x = np.arange(n+1)
y = binom.pmf(k=x,n=n,p=p_0)

plt.bar(x,y, width=.5)



x_o = np.arange(binom.ppf(q=0.95, n=n, p=p_0), n+1)
y_o = binom.pmf(k=x_o, n=n, p=p_0)
plt.bar(x_o, y_o, color= "red", width=.5);
```

## Macht, Fehler 2. Art

$\pi_A=.8$
