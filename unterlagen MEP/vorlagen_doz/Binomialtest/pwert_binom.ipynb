# $p$-Wert

## Einseitig (nach oben)

Wir beginnen mit einer Binomialverteilung mit $n=100$ und $\pi=0.5$

```python
%matplotlib inline

from scipy.stats import binom, binomtest 
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (15,5)
import numpy as np
n = 100
p_0 = 0.5
```

### Beobachtung $k$

```python
k = 60

x = np.arange(n+1)
y = binom.pmf(k=x,n=n,p=p_0)

plt.bar(x,y, width=.5)

p_value = binomtest(k=k,n=n,p=p_0, alternative="greater")

print("p-Wert:",p_value.pvalue)

x_o = np.arange(k, n+1)
y_o = binom.pmf(k=x_o, n=n, p=p_0)
plt.bar(x_o, y_o, color= "red", width=.5);
```

## Zweiseitig
### Beobachtung $k$

```python
k = 60

x = np.arange(n+1)
y = binom.pmf(k=x,n=n,p=p_0)

plt.bar(x,y, width=.5)

p_value = binomtest(k=k,n=n,p=p_0, alternative="two-sided")

print("p-Wert:",p_value.pvalue)

x_o = np.arange(k, n+1)
y_o = binom.pmf(k=x_o, n=n, p=p_0)
x_u = np.arange(0, n-( k-1))
y_u = binom.pmf(k=x_u, n=n, p=p_0)

plt.bar(x_u, y_u, color= "red", width=.5)
plt.bar(x_o, y_o, color= "red", width=.5)
;
```
