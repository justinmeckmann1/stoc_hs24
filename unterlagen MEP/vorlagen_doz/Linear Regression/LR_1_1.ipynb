# Beispiel 1.1

Wir werden als Statistiker von einer Firma mit dem Auftrag betraut, eine Analyse und Strategie auszuarbeiten, wie der Verkauf eines bestimmten Produktes dieser Firma gesteigert werden kann. Die Firma stellt uns die Daten von Werbebudget und Verkauf zur Verfügung.  Der Datensatz `Werbung` besteht aus dem `Verkauf` dieses Produktes in 200 verschiedenen Märkten und den Werbebudgets für dieses Produkt in diesen Märkten für die drei verschiedenen Medien `TV`, `Radio` und `Zeitung`. 

```python
import pandas as pd

werbung = pd.read_csv("./Daten/Werbung.csv").drop(["Unnamed: 0"], axis=1)

werbung.head()
print(werbung.head())
```

Dabei sind `TV`, `Radio` und `Zeitung` die Werbeausgaben in Tausenden CHF und der `Verkauf` die verkauften Einheiten in Tausenden CHF. In Markt 1 wurden beispielsweise CHF 230100 in TV-,  CHF 37800
in Radio- und CHF 69200 in Zeitungswerbung investiert. Dabei wurden 22100 Einheiten des Produktes verkauft.

Wir können nun für jedes einzelne Werbebudget ein Streudiagrammen mit `Verkauf` erstellen (siehe Abbildung unten).

```python
import matplotlib.pyplot as plt

# Create the first subplot with the y-axis label "Verkauf"
plt.subplot(131)
werbung.plot(kind="scatter", x="TV", y="Verkauf", ax=plt.gca())
plt.ylabel("Verkauf")  # Display the label on the first subplot

# Create the second subplot without the y-axis label
plt.subplot(132)
werbung.plot(kind="scatter", x="Radio", y="Verkauf", ax=plt.gca())
plt.ylabel("")  # Remove the y-axis label for the second subplot

# Create the third subplot without the y-axis label
plt.subplot(133)
werbung.plot(kind="scatter", x="Zeitung", y="Verkauf", ax=plt.gca())
plt.ylabel("")  # Remove the y-axis label for the third subplot

# Display the plot
plt.show()
```
