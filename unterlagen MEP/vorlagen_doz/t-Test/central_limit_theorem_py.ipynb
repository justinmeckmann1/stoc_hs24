```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

werte = np.array([0, 10, 11])
ew = np.sum(werte * 1/3)
ew
```

```python
var_X = np.sum((werte - ew)**2 * 1/3)

var_X
```

```python
sd_X = np.sqrt(var_X)
sd_X
```

```python
# Zieht 10-mal aus der Menge {0,10,11} einen Wert mit gleicher W'keit 

sim = np.random.choice(werte, 10, replace = True)

#Vektor mit 10 Werten
sim
```

```python
n = 16
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,1)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.5), edgecolor="black")

n = 64
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,2)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.25), edgecolor="black")

n = 256
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,3)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.1), edgecolor="black")

n = 1024
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,4)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.1), edgecolor="black");
```

```python
from scipy.stats import norm
x = np.linspace(0,13,1000)
n = 16
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,1)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.5), edgecolor="black", density=True)
plt.plot(x, norm.pdf(x, loc=7, scale=sd_X/np.sqrt(n)), linewidth=2)

n = 64
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,2)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.25), edgecolor="black", density=True)
plt.plot(x, norm.pdf(x, loc=7, scale=sd_X/np.sqrt(n)), linewidth=2)

n = 256
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,3)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.1), edgecolor="black", density=True)
plt.plot(x, norm.pdf(x, loc=7, scale=sd_X/np.sqrt(n)), linewidth=2)

n = 1024
sim = np.random.choice(werte, n*1000, replace = True).reshape(n,1000)
plt.subplot (2,2,4)
plt.hist(np.mean(sim,axis=0), bins=np.arange(13, step=.1), edgecolor="black", density=True)
plt.plot(x, norm.pdf(x, loc=7, scale=sd_X/np.sqrt(n)), linewidth=2);
```
