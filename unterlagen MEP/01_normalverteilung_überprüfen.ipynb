```python
# 1) Visuell: Q-Q Plot --> Normalverteilung ist gerechtfertigt, wenn Daten ungefähr der Geraden folgen
# Im gezeigten Beispiel ist eine Normalverteilung gerechtfertigt. Da der Datensatz jedoch sehr klein ist,
# ist eine abschliessendes Beurteilung schwierig. 

import scipy.stats as st
import matplotlib.pyplot as plt

data = [71, 69, 67, 68, 73, 72, 71, 71, 68, 72, 69, 72]
st.probplot(data, plot=plt)
plt.show()
```
