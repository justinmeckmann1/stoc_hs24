```python
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
# Plot normal distribution

loc     = 70  # 𝜇 Mittelwert
scale   = 1.5  # 𝜎 Standardabweichung 	

# Generate data for normal distribution
x = np.linspace(loc - 3*scale, loc + 3*scale, 100)
normal_dist = st.norm.pdf(x, loc, scale)

# plt.figure(figsize=(5, 5))
plt.plot(x, normal_dist, label='Normalverteilung')
plt.title('Normalverteilung')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# Parameters for the distributions
n = 10  # Anzahl versuche (trails) / durchführung
p = 0.5 # Erfolgswahrscheinlichkeit 

binomial_dist = st.binom.pmf(np.arange(0, n+1), n, p)

# plt.figure(figsize=(5, 5))
plt.bar(np.arange(0, n+1), binomial_dist)
plt.title('Binomialverteilung ')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend(['Binomialverteilung '])
plt.show()
```
