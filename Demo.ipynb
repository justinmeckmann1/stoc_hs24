```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameter
n = 10  # Anzahl der Versuche
p = 0.8  # Erfolgswahrscheinlichkeit

# Wertebereich für die Zufallsvariable
k = np.arange(0, n+1)

# Wahrscheinlichkeiten berechnen
pmf = binom.pmf(k, n, p)

# Plot erstellen
plt.bar(x=k, height=pmf, label='Binomialverteilung')
plt.title("Diskrete Binomialverteilung")
plt.xlabel("Anzahl Erfolge (k)")
plt.ylabel("Wahrscheinlichkeit")
plt.xticks(k)
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Daten
data = [2, 3, 3, 5, 6]
n = len(data)

# Sortierte Daten
sorted_data = np.sort(data)

# Werte für die EVF
x = np.concatenate([[-np.inf], sorted_data, [np.inf]])
y = np.arange(0, n+2) / n  # Anpassen der Dimension von y

# Plot
plt.step(x, y, where="post", label="Empirische Verteilungsfunktion (EVF)")
plt.title("Empirische Verteilungsfunktion")
plt.xlabel("x")
plt.ylabel("F_n(x)")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameter der Normalverteilung
mu = 170  # Mittelwert
sigma = 10  # Standardabweichung

# Wertebereich
x = np.linspace(140, 200, 1000)

# Dichtefunktion (PDF)
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Verteilungsfunktion (CDF) für 160 und 180
cdf_160 = norm.cdf(160, loc=mu, scale=sigma)
cdf_180 = norm.cdf(180, loc=mu, scale=sigma)

# Wahrscheinlichkeit berechnen
probability = cdf_180 - cdf_160

# Plot der Dichtefunktion
plt.plot(x, pdf, label="Normalverteilung")
plt.fill_between(x, 0, pdf, where=(x >= 160) & (x <= 180), color='gray', alpha=0.5,
                 label=f"P(160 ≤ X ≤ 180) = {probability:.2f}")
plt.title("Normalverteilung")
plt.xlabel("x")
plt.ylabel("Dichte")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameter der Exponentialverteilung
lambda_rate = 2  # Rate (Ereignisse pro Minute)

# Wertebereich
x = np.linspace(0, 3, 1000)

# Wahrscheinlichkeitsdichtefunktion (PDF)
pdf = expon.pdf(x, scale=1/lambda_rate)

# Verteilungsfunktion (CDF)
cdf = expon.cdf(x, scale=1/lambda_rate)

# Plot
plt.figure(figsize=(10, 6))

# PDF
plt.plot(x, pdf, label="Wahrscheinlichkeitsdichtefunktion (PDF)")
# CDF
plt.plot(x, cdf, label="Verteilungsfunktion (CDF)", linestyle="--")

plt.title("Exponentialverteilung")
plt.xlabel("x")
plt.ylabel("Dichte / Wahrscheinlichkeit")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameter der Exponentialverteilung
lambda_rate = 0.5  # Rate (Ausfälle pro Stunde)

# Wertebereich
x = np.linspace(0, 10, 1000)

# Wahrscheinlichkeitsdichtefunktion (PDF)
pdf = expon.pdf(x, scale=1/lambda_rate)

# Plot
plt.plot(x, pdf, label=r"$f(x) = \lambda e^{-\lambda x}$, $\lambda = 0.5$")
plt.title("Wahrscheinlichkeitsdichtefunktion der Exponentialverteilung")
plt.xlabel("x (Stunden)")
plt.ylabel("Dichte")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation: Würfelwürfe
n = 50  # Anzahl der Würfe pro Stichprobe
samples = 10000  # Anzahl der Stichproben

# Zufallszahlen simulieren
data = np.random.randint(1, 7, (samples, n))  # Würfelzahlen 1 bis 6
means = data.mean(axis=1)  # Mittelwert jeder Stichprobe

# Histogramm der Mittelwerte
plt.hist(means, bins=30, density=True, alpha=0.6, color='gray', label="Stichprobenmittelwerte")

# Normalverteilung plotten
mu = 3.5
sigma = np.sqrt(35/12 / n)
x = np.linspace(2.5, 4.5, 10000)
plt.plot(x, (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2),
         label="Normalverteilung", color='red')

plt.title("Zentraler Grenzwertsatz")
plt.xlabel("Mittelwert")
plt.ylabel("Dichte")
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import norm

# Gegebene Werte
mu_0 = 500  # Populationsmittelwert
x_bar = 495  # Stichprobenmittelwert
sigma = 10  # Populationsstandardabweichung
n = 30  # Stichprobengröße
alpha = 0.05  # Signifikanzniveau

# z-Wert berechnen
z = (x_bar - mu_0) / (sigma / (n ** 0.5))

# Kritischer z-Wert
z_critical = norm.ppf(1 - alpha / 2)  # Zweiseitiger Test

# Entscheidung
if abs(z) > z_critical:
    print(f"z = {z:.2f} > z_critical = {z_critical:.2f}. Nullhypothese ablehnen.")
else:
    print(f"z = {z:.2f} <= z_critical = {z_critical:.2f}. Nullhypothese beibehalten.")
```

```python
from scipy.stats import bernoulli

# Parameter p
p = 0.6

# Wahrscheinlichkeit für x = 1 (Erfolg)
probability = bernoulli.pmf(1, p)
print(f"P(X = 1) = {probability}")

# Wahrscheinlichkeit für x = 0 (Misserfolg)
probability_0 = bernoulli.pmf(0, p)
print(f"P(X = 0) = {probability_0}")
```
