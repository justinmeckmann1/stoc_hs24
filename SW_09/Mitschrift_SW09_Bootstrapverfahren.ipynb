## Bootstrapverfahren

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([30,37,36,43,42,43,43,46,41,42])
n = x.size

print(x)
print(f"Mittelwert: {np.mean(x)}")

plt.hist(x, density=True)
plt.show()
```

```python
nboot = 20 # Anzahl der Wiederholungen
tmpdata = np.random.choice(x, n*nboot, replace=True)

bootstrapsample = np.reshape(tmpdata, (nboot, n))
# print(bootstrapsample)

xbarstar = bootstrapsample.mean(axis=1) # Mittelwerte über die "Zeilen"
print(xbarstar)

# print(tmpdata)
# print(f"Mittelwert: {np.mean(tmpdata)}")
```

```python
# Vertrauensintervall

alpha = 0.05

I = np.quantile(xbarstar, q=[alpha/2, 1-alpha/2])
print(f'\nVertraunsinterval: {I}')
```

```python
# np.random.seed(8)
x = np.random.normal(loc=40, scale=5, size=10000)
x = np.reshape(x,(100,100))
nboot = 1000
n = 100
k = 0

for i in range(0,100):
    tmpdata = np.random.choice(x[i], n*nboot, replace=True)
    bootstrapsample = np.reshape(tmpdata, (nboot, n))
    xbarstar = bootstrapsample.mean(axis=1)
    d = np.percentile(xbarstar, q=[2.5,97.5])
    ci = np.percentile(xbarstar, q=[2.5,97.5])
    plt.plot([i,i], ci)
    if d[0] <= 40 <= d[1]:
        k += 1

print(f'Der wahre Parameter 40 lag {k} mal im Vertrauensintervall')
```
