# Aufgaben SW 2

a)      
$P(K)$: Wahrscheinlichkeit, dass eine Person krank ist      
$P(\bar{T})$: Wahrscheinlichkeit, dass eine Person negativ getestet wird        
$P(K|T)$: Wahrscheinlichkeit, dass eine positiv getestete Person krank ist       
$P(T|K)$: Wahrscheinlichkeit, dass eine kranke Person positiv getestet wird      
$P(\bar{T}|\bar{K})$: Wahrscheinlichkeit, dass eine gesunde Person negativ getestet wird     

b)      
$P(K)$: 1%
$P(\bar{T})$: 3% von 99% + 2% von 1% = 2.97% + 0.02% = 3.99%
$P(K|T)$: 98%
$P(T|K)$: 98%
$P(\bar{T}|\bar{K})$: 97%
