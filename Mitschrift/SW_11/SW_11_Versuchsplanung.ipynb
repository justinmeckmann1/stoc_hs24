# SW11 - Versuchsplanung

### Parameterschätzung
<u>Modell:</u> 2-Weg-Varianzanalyse ohne Wechselwirkung     
$Y_{ijk} = \mu + \alpha_i + \beta_j + \epsilon_{ijk}$

__Nebenbedingungen:__       
$\sum_{i=1}^{a}\alpha_i = 0$        
$\sum_{j=1}^{b}\beta_j = 0$     

__z.B.:__
A,B, zwei Stufen        
$\alpha_1 + \alpha_2 = 0$ --> $\alpha_1 = -\alpha_2$                
$\beta_1 + \beta_2 = 0$ --> $\beta_1 = -\beta_2$        

__Parameterschätzung:__         
$\mu = \frac{1}{n} \sum_{i,j,k}Y_{ijk} = \bar{Y} ...$     
$\mu = a*b*w$ mit w= Anzahl der Wiederholungen      
$\bar{\alpha}_i = \bar{Y}_{i..} - \bar{Y}$  [$\alpha_i = \bar{\mu}_i.. - \mu$]
